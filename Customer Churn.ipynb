# Converting 0 and 1 Values of SeniorCitizen to Yes/No to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
        
```

# Lets make graphas

# Count of the Customer by churn

```python
plt.figure(figsize = (3,4))
ax = sns.countplot(x = 'Churn',data = df)
ax.bar_label(ax.containers[0])
plt.title('Count of the Customer by churn')
plt.show()
```

# Here we know more customer are not churned

# make group between Count of the Customer by churn and churn customer

```python
gb = df.groupby('Churn').agg({'Churn':'count'})
gb
```

# Percentage of churned customer

```python
plt.figure(figsize = (3,4))
plt.pie(gb['Churn'], labels = gb.index,autopct = '%1.2f%%')
plt.title("Percentage of churned customer",fontsize = 10)
plt.show()
```

# creating subs-plots

```python
# List of columns you want to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid (3 rows x 3 columns)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(13, 18))
fig.suptitle("Count Plots of Service Features", fontsize=20)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create count plots and add bar labels
for i, col in enumerate(cols):
    ax = sns.countplot(data=df, x=col, hue=col, ax=axes[i], palette="Set2", legend=False)
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

    # Add labels to each bar in the plot
    for container in ax.containers:
        ax.bar_label(container, fontsize=8)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# "Each subplot shows the count of categories with bars representing the frequency, and numerical labels are added on top of each bar."

```python
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(11, 18))
fig.suptitle("Count Plots of Service Features", fontsize=20)

axes = axes.flatten()

for i, col in enumerate(cols):
    ax = sns.countplot(data=df, x=col, hue='Churn', ax=axes[i], palette="Set2", legend=False)
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

    for container in ax.containers:
        ax.bar_label(container, fontsize=8)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# "Each subplot displays counts of service feature categories with bars color-coded by churn status, and frequency labels shown on top of each bar."
